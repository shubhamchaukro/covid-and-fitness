```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score,mean_squared_error    
from sklearn.model_selection import train_test_split
```

```python
skipped=skipped.fillna(0)
skipped['weightchange']=skipped['gain']-skipped['loss']

#skipped.loc[skipped.weightchange==max(skipped.weightchange)]
#skipped=skipped.loc[70].drop
#skipped=skipped.drop(skipped.loc[skipped.weightchange==max(skipped.weightchange)].index)
#skipped.describe()
skipped=skipped.drop(['gain','loss'],axis=1)
skipped.index = (np.arange(1,len(skipped)+1,1))
skipped
```

```python
X=skipped.iloc[:,1:22]
y=skipped.iloc[:,22]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state = 3)
```

```python
y_train
```

```python
clf = GradientBoostingRegressor(n_estimators=10,max_features=7,random_state=0)
clf.fit(X_train,y_train)
```

```python
pred=clf.predict(X_test)
#acc = accuracy_score(y_test,pred)
mse = mean_squared_error(y_test,pred)
mse
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.close('all')
```

```python

fig, ax = plt.subplots()
bar_plot=plt.bar(range(len(names)),clf.feature_importances_)
plt.yticks(np.linspace(0,0.3,10),fontsize=12)
plt.xticks(range(len(names)),fontsize=12)
plt.xlabel("Features",fontsize=15)
plt.ylabel("Importance",fontsize=15)
plt.title("Feature Importance",fontsize=18)
def autolabel(rects):
    for idx,rect in enumerate(bar_plot):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 0.4*height,
                names[idx],
                ha='center', va='bottom', rotation=90, fontsize=14)
autolabel(bar_plot)
fig.set_figheight(5)
fig.set_figwidth(10)
plt.show()
```
